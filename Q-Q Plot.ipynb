# Q-Q Plot
- Also known as **Quartile-Quartile Plots**
- Applications of Q-Q Plot
    - Find out if a random distribution is Gaussian or not
    - Find out if two random distributions have a similar spread or not
    
- Calculates the quartiles of each given distribution and compares them to check similarities between the two

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# plot N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)
'''
it says that generate a normal distribution(np.random.normal()) with a mean 0 (loc = 0) and 
standard deviation = 1 (scale = 1) and generate 1000 such samples and store them in 'std_normal')
where 
    loc = 0 is mean
    scale = 1 is standard deviation
'''

# 0 to 100th percentile of std_normal
for i in range(0,101):
    print(i, np.percentile(std_normal, i)) # gives the 1st percentile, 2nd percentile ... till 100th percetile for std_normal
    
# refer : https://docs.scipy.org/doc/numpy/reference/generated/numpy.percentile.html
```

```python
# generate 100 samples from N(20,5)
measurements = np.random.normal(loc=20, scale=5, size=100)

# create Q-Q plot
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

- Create a Q-Q plot (stats.probplot) for 'measurements' and compare it against a standard normal distribution (dist='norm').
- For plotting use the library 'pylab'

```python
# now generate 1000 samples from N(20,5)
measurements = np.random.normal(loc=20, scale=5, size=1000)

# create Q-Q plot
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

- More points align to the straight line

```python
# now generate 50000 samples from N(20,5)
measurements = np.random.normal(loc=20, scale=5, size=50000)

# create Q-Q plot
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

- As we keep increasing the number of points, more and more number of points start aligning with the straight line

```python
# now generate 50 samples from N(20,5)
measurements = np.random.normal(loc=20, scale=5, size=50)

# create Q-Q plot
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

- And as we decrease the number of points, they start mis-aligning themselves from the straight line
- Which is a **limitation** of the Q-Q plot, if the number of samples is small, it becomes very difficult to read the Q-Q plot

### Let's plot a Q-Q plot comparing a Uniform distribution with a Normal distribution

```python
# now generate 100 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=-1, size=100)

# create Q-Q plot
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```
